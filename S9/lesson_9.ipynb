```python
def rule_rating(row):
    if row["Rating"] >=4:
        return "High"
    elif 2 < row["Rating"] < 4:
        return "Medium"
    else:
        return "Low"
        
```

```python
data["Rating_label"] = data.apply(rule_rating, axis=1)
```
