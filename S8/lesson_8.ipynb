```python
from scipy import stats
from scipy.stats.mstats import winsorize
```

```python
z_scores = stats.zscore(df['Installs'])
treshold = 3
overvalues = df['Installs'][z_scores > treshold]
overvalues
```

```python
df.loc[z_scores > treshold, 'Installs'] = df['Installs'].median()
```

```python
plt.hist(df['Installs'])
plt.show()
```

```python
z_scores = stats.zscore(df['Reviews'])
treshold = 3
overvalues = df['Reviews'][z_scores > treshold]
overvalues
```

```python
df.loc[z_scores > treshold, 'Reviews'] = df['Reviews'].median()
```

```python
plt.hist(df['Reviews'])
plt.show()
```

```python
df['Rating'] = winsorize(df['Rating'], limits=[0.05, 0.05])
```

```python
plt.hist(df['Rating'])
plt.show()
```
